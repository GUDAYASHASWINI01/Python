```python
def sum_of_even_numbers(n):
  """
  Calculates the sum of all even numbers between 1 and n (inclusive).

  Args:
    n: A positive integer.

  Returns:
    The sum of even numbers, or 0 if n is not a positive integer.
  """
  if not isinstance(n, int) or n <= 0:
    return 0

  sum_of_evens = 0
  for i in range(2, n + 1, 2):
    sum_of_evens += i
  return sum_of_evens

# Get input from the user
n = int(input("Enter a positive integer: "))

# Calculate and print the sum of even numbers
result = sum_of_even_numbers(n)
print(f"The sum of even numbers between 1 and {n} is: {result}")
```
